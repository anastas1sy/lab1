#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_glasnie(text):
    glasnie = 'aeiou'
    count = 0
    for i in text.lower():
        if i in glasnie:
            count +=1
    return count 

def test_count_glasnie():

    assert count_glasnie("hello world") == 3
    assert count_glasnie("HELLO WORLD") == 3
    assert count_glasnie("Hello World") == 3
    assert count_glasnie("bcdfg") == 0
    assert count_glasnie("aeiou") == 5
    assert count_glasnie("") == 0
    assert count_glasnie("aaaeeeiiiooouuu") == 15
    assert count_glasnie("h3ll0 w0rld!") == 0
    assert count_glasnie("yesterday") == 3
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_count_glasnie()
    
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def unique_symbol(text):
    symbol_count = {}
    for char in text:
        if char in symbol_count:
            return False  
        symbol_count[char] = 1
    return True

def text_unique_symbol():
    
    assert unique_symbol('abcdaa') == False 
    assert unique_symbol('abcde') == True
    assert unique_symbol('') == True
    assert unique_symbol('b') == True
    assert unique_symbol('a b!c') == True
    assert unique_symbol('a b ') == False
    assert unique_symbol('1 2?4') == True
    assert unique_symbol('?!%^') == True
    assert unique_symbol('1231') == False

if __name__ == '__main__':
    text_unique_symbol()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def binary_number(number):

    if number is None:
        return "Функция не принимает пустое значение"

    if number < 0:
        return "Функция работает только с положительными числами."
    
    bin_number = bin(number)[2:]

    return bin_number.count('1')

def test_binary_number():
    assert binary_number(0) == 0
    assert binary_number(255) == 8
    assert binary_number(-3) == "Функция работает только с положительными числами."
    assert binary_number(None) == "Функция не принимает пустое значение"

if __name__ == '__main__':
    test_binary_number()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):  
    if n < 0:
        raise ValueError("Число должно быть положительным")
    
    if n < 10:
        return 0
    
    steps = 0
    current = n
    
    while current >= 10:
        product = 1
        for digit in str(current):
            product *= int(digit)
        
        current = product
        steps += 1
    
    return steps

def test_magic():
    
    assert magic(39) == 3
    assert magic(4) == 0
    assert magic(999) == 4
    assert magic(25) == 2
    assert magic(10) == 1
    assert magic(6788) == 6
    assert magic(10) == 1
    assert magic(1) == 0
    assert magic(0) == 0

if __name__ == "__main__":
    test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import math
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError('Векторы должны быть одинаковой длины')
    
    if len(pred) == 0:
        raise ValueError('Векторы не могут быть пустыми')
    n = len(pred)
    sum_os = 0.0

    for i in range(n):
        sum_os += (pred[i] -  true[i]) ** 2
    
    mse = math.sqrt(sum_os / n)
    return mse

def test_mse():

    vector1 = [1, 2, 3, 4, 5]
    vector2 = [1, 2, 3, 4, 5]
    assert mse(vector1, vector2) == 0.0
    
    vector1 = [1, 2, 3]
    vector2 = [4, 5, 6]
    expected = 3.0
    result = mse(vector1, vector2)
    assert abs(result - expected) < 1e-10
    
    vector1 = [-1, -2, -3]
    vector2 = [1, 2, 3]
    expected = math.sqrt(56 / 3)
    result = mse(vector1, vector2)
    assert abs(result - expected) < 1e-10

    vector1 = [5]
    vector2 = [3]
    expected = 2.0
    result = mse(vector1, vector2)
    assert abs(result - expected) < 1e-10

    vector1 = [1.5, 2.5, 3.5]
    vector2 = [0.5, 1.5, 2.5]
    expected = 1.0
    result = mse(vector1, vector2)
    assert abs(result - expected) < 1e-10

    try:
        mse([1, 2], [1, 2, 3])
        assert False
    except ValueError:
        pass
    
    try:
        mse([], [])
        assert False
    except ValueError:
        pass 
    

    vector1 = [1, 3, 5, 7, 9]
    vector2 = [2, 4, 6, 8, 10]
    result1 = mse(vector1, vector2)
    result2 = mse(vector1, vector2)
    assert abs(result1 - result2) < 1e-10

if __name__ == "__main__":
    test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    
    if n == 1:
        return "(1)"
    
    factors = {}
    temp = n
    divisor = 2
    
    while divisor * divisor <= temp:
        while temp % divisor == 0:
            factors[divisor] = factors.get(divisor, 0) + 1
            temp //= divisor
        divisor += 1
    
    if temp > 1:
        factors[temp] = factors.get(temp, 0) + 1
    
    result_parts = []
    for prime in sorted(factors.keys()):
        power = factors[prime]
        if power == 1:
            result_parts.append(f"({prime})")
        else:
            result_parts.append(f"({prime}**{power})")
    
    return "".join(result_parts)

def test_prime_factors():
    
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(2) == "(2)"
    assert prime_factors(17) == "(17)"
    assert prime_factors(13) == "(13)"
    assert prime_factors(1) == "(1)"
    assert prime_factors(4) == "(2**2)"
    assert prime_factors(8) == "(2**3)"
    assert prime_factors(9) == "(3**2)"
    assert prime_factors(12) == "(2**2)(3)"
    assert prime_factors(18) == "(2)(3**2)"
    assert prime_factors(30) == "(2)(3)(5)"
    assert prime_factors(100) == "(2**2)(5**2)", "Тест 12 не пройден"
    assert prime_factors(2310) == "(2)(3)(5)(7)(11)"
    assert prime_factors(72) == "(2**3)(3**2)"

    try:
        prime_factors(0)
        assert False
    except ValueError:
        pass
    
    try:
        prime_factors(-5)
        assert False
    except ValueError:
        pass
    
if __name__ == "__main__":
    test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    
    if number <= 0:
        return 'It is impossible'

    k = 1
    total_cubes = 0
    
    while total_cubes < number:
        total_cubes += k * k
        if total_cubes == number:
            return k
        k += 1
    
    return "It is impossible"

def test_pyramid():

    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(2) == "It is impossible"
    assert pyramid(3) == "It is impossible"
    assert pyramid(6) == "It is impossible"
    assert pyramid(15) == "It is impossible"
    assert pyramid(29) == "It is impossible"    
    assert pyramid(0) == "It is impossible"
    assert pyramid(-5) == "It is impossible"
    assert pyramid(506) == 11

if __name__ == "__main__":
    test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def balanced_number(n):

    if n <= 0:
        raise ValueError("Число должно быть положительным")
    
    digits = [int(d) for d in str(n)]
    length = len(digits)
    
    if length <= 2:
        return False
    
    if length % 2 == 1:
        mid = length // 2
        left_sum = sum(digits[:mid])
        right_sum = sum(digits[mid + 1:])
    else:
        mid = length // 2
        left_sum = sum(digits[:mid - 1])
        right_sum = sum(digits[mid + 1:])
    
    return left_sum == right_sum

def test_balanced_number():

    assert balanced_number(121) == True
    assert balanced_number(12321) == True
    assert balanced_number(13531) == True
    assert balanced_number(1234) == False
    assert balanced_number(1221) == True
    assert balanced_number(123321) == True
    assert balanced_number(123) == False
    assert balanced_number(12345) == False
    assert balanced_number(1222) == False
    assert balanced_number(100) == False
    assert balanced_number(999) == True
    assert balanced_number(1) == False
    assert balanced_number(12) == False
    assert balanced_number(99) == False
    assert balanced_number(1234554321) == True
    assert balanced_number(12345654321) == True
    try:
        balanced_number(0)
        assert False
    except ValueError:
        pass
    
    try:
        balanced_number(-123)
        assert False
    except ValueError:
        pass

if __name__ == "__main__":
    test_balanced_number()
```
